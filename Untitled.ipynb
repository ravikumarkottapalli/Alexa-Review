```python
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
sent = SentimentIntensityAnalyzer()
```

```python
def feedback(text):
    if not isinstance(text, str):
        return "unknown"
    score =sent.polarity_scores(text)['compound']
    if score>=0.1:
        return "positive"
    elif score<= (-0.1):
        return "Negative"
    else:
        return "Neautral" 
```

```python
df1['verified_reviews'].apply(feedback)
```

```python
df1['Result'] = df1['verified_reviews'].apply(feedback).to_frame()
```
